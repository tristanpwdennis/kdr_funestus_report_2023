```python
from vina import Vina
import itertools as it


receptors = ['AFUN2_008728.R15290.L976F.1.pdbqt','AFUN2_008728.R15290.WT.1.pdbqt','AGAP004707-RD.995F.1.pdbqt','AGAP004707-RD.995S.1.pdbqt','AGAP004707-RD.WT.pdbqt']
ligands = ['permethrin', 'deltamethrin', 'ddt']

for seed in [53867, 12345, 75436, 345634, 95347,84653]:
    for a,b in it.product(receptors, ligands):
        v = Vina(sf_name='vina', seed=seed)    
        v.set_receptor(a)
        v.set_ligand_from_file(f'../ligand_files/ligand_{b}.pdbqt')
        v.compute_vina_maps(center=[-0.240, 0.870, 5.462], box_size=[20, 20, 20])
        energy = v.score()
        print('Score before minimization: %.3f (kcal/mol)' % energy[0])
        # Minimized locally the current pose
        energy_minimized = v.optimize()
        print('Score after minimization : %.3f (kcal/mol)' % energy_minimized[0])
        v.dock(exhaustiveness=32, n_poses=20)
        v.write_poses(f'{a}.{b}.{seed}.pdbqt', n_poses=5, overwrite=True)
```
